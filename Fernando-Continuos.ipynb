# Fórmulas para los calculos necesarios.

## Frecuencia Absoluta
La frecuencia absoluta es simplemente el conteo de cuántas veces aparece un valor o un rango de valores en un conjunto de datos.

Fórmula:

$$ Frecuencia\ Absoluta =Numero\ de\ ocurrencias\ de\ un\ valor\ o\ intervalo $$

## Frecuencia Relativa
La frecuencia relativa es la proporción de veces que ocurre un valor o intervalo en relación al total de observaciones. Se expresa como una fracción o porcentaje, y permite comparar la importancia de cada categoría dentro del conjunto de datos.

Fórmula:

$$ Frecuencia\ Relativa = \frac{Total\ de\ Observaciones}{Frecuencia\ Absoluta} $$

## Frecuencia Acumulada
La frecuencia acumulada es la suma de las frecuencias absolutas de una categoría y todas las categorías anteriores. Muestra el total de observaciones que caen dentro de un intervalo hasta un punto específico, facilitando el análisis de datos en rangos acumulativos.

Fórmula: 

$$ Frecuencia\ Acumulada(i)= \sum^{i}_{j=1}Frecuencia\ Absoluta(j) $$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos de la tabla
intervals = ['0.0 - 100.0', '100.1 - 200.0', '200.1 - 300.0', '300.1 - 400.0', '400.1 - 500.0']
frequencies = [30.5, 85.2, 150.0, 120.3, 114.0]

# Crear un DataFrame
df = pd.DataFrame({'Intervalo': intervals, 'Frecuencia': frequencies})

# Frecuencia acumulada
df['Frecuencia Acumulada'] = df['Frecuencia'].cumsum()

print("Setup terminado")
```

```python
plt.figure(figsize = (12, 6)) ## (ancho, alto) de la gráfica 

plt.bar(intervals, frequencies, # X y Y en la gráfica
        width = 1, # Ancho de las barras
        edgecolor = "k", # Color del borde de barra
        color = ["#736e6e", "#736e6e", "#00ff11", "#736e6e", "#736e6e"]) # Colores de las barras
plt.xlim(-0.5, len(intervals)) # Intervalo escala en x
plt.ylim(0, max(frequencies) + 10) # Intervalo escala en y
plt.grid(True) # Cuadricular gráfica
plt.title("Histograma", fontsize=20) # Título de la gráfica
plt.xlabel('Intervalo', fontsize=15) # Etiqueta en x
plt.xticks(rotation=45)
plt.ylabel('Frecuencia absoluta', fontsize=15) # Etiqueta en y
plt.grid()
plt.tight_layout()
plt.show() # Imprime la gráfica
```

```python
plt.figure(figsize = (12, 6)) # Ancho, largo del grafico

# Diagrama de barras con:
# - Barras al 80% (height)
# - Contorno negro ( edgecolor = "k")
plt.barh(intervals, frequencies,
        height = 0.8, edgecolor = "k", 
        color = ["#736e6e", "#736e6e", "#00ff11", "#736e6e", "#736e6e"])
plt.yticks(intervals)

plt.title("Diagrama de barras") # Titulo
plt.xlabel("Frecuencia absoluta") # Etiqueta eje X
plt.ylabel("Intervalo") # Etiqueta eje Y
plt.grid(axis='x')
plt.tight_layout()
plt.show()#Se muestra gráfica en pantalla
```

```python
plt.figure(figsize=(12, 6))  # Ancho, largo del gráfico
frequencies_relative = df['Frecuencia'] / df['Frecuencia'].sum()

# Crear listas para incluir 0 al inicio y al final
intervals_extended = ['0.0 - 100.0'] + df['Intervalo'].tolist() + ['400.1 - 500.0']
frequencies_relative_extended = [0] + frequencies_relative.tolist() + [0]

# Polígono de frecuencia
plt.plot(intervals_extended, frequencies_relative_extended,
         linewidth=5, color="r", linestyle=":",
         marker="v", markersize=15, markeredgecolor="b", markerfacecolor="y")

plt.xticks(rotation=45)  # Cambio de las marcas del eje x
plt.xlim(-0.1, len(df))  # Limite eje X ajustado
plt.title("Polígono de Frecuencia")  # Título
plt.xlabel("Intervalo")  # Etiqueta eje X
plt.ylabel("Frecuencia relativa")  # Etiqueta eje Y
plt.grid(True)  # Cuadricula
plt.show()  # Se muestra gráfica en pantalla
```

```python
separaciones = [0, 0, 0.1, 0, 0]
plt.figure(figsize=(8, 5))
plt.pie(frequencies_relative[::-1],  # Invertir el orden de los datos
        explode=separaciones[::-1],  # Invertir las separaciones si es necesario
        labels=df['Intervalo'][::-1],  # Invertir las etiquetas
        autopct='%1.1f%%',
        startangle=90)  # Mantener el inicio en 12h
plt.title('Diagrama de Pastel de Frecuencia Relativa')
plt.tight_layout()
plt.show()
```

```python
# 5. Ojiva de Frecuencia Acumulada
plt.figure(figsize=(8, 5))

# Calcular la frecuencia acumulada
df['Frecuencia Acumulada'] = df['Frecuencia'].cumsum()

# Calcular porcentajes de la frecuencia acumulada
frecuencia_acumulada_porcentajes = df['Frecuencia Acumulada'] / df['Frecuencia Acumulada'].iloc[-1] * 100

plt.plot(df['Intervalo'], frecuencia_acumulada_porcentajes, marker='o', linestyle='-', color='green')
plt.title('Ojiva de Frecuencia Acumulada')
plt.xlabel('Intervalo')
plt.ylabel('Frecuencia Acumulada (%)')  # Etiqueta en porcentaje
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```
