```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos de la tabla
intervals = ['0.0 - 100.0', '100.1 - 200.0', '200.1 - 300.0', '300.1 - 400.0', '400.1 - 500.0']
frequencies = [30.5, 85.2, 150.0, 120.3, 114.0]

# Crear un DataFrame
df = pd.DataFrame({'Intervalo': intervals, 'Frecuencia': frequencies})

# Frecuencia acumulada
df['Frecuencia Acumulada'] = df['Frecuencia'].cumsum()

print("Setup terminado")
```

```python
# 1. Histograma de Frecuencia Absoluta
plt.figure(figsize=(8, 5))
plt.hist(np.repeat(intervals, frequencies), bins=len(intervals), edgecolor='black', alpha=0.7)
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Intervalo')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 2. Diagrama de Barras Horizontales
plt.figure(figsize=(8, 5))
plt.barh(df['Intervalo'], df['Frecuencia'], color='skyblue', edgecolor='black')
plt.title('Diagrama de Barras de Frecuencia Absoluta')
plt.xlabel('Frecuencia Absoluta')
plt.ylabel('Intervalo')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# 3. Polígono de Frecuencia Relativa
frequencies_relative = df['Frecuencia'] / df['Frecuencia'].sum()
plt.figure(figsize=(8, 5))
plt.plot(df['Intervalo'], frequencies_relative, marker='o', linestyle='-', color='orange')
plt.title('Polígono de Frecuencia Relativa')
plt.xlabel('Intervalo')
plt.ylabel('Frecuencia Relativa')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 4. Diagrama de Pastel de Frecuencia Relativa
plt.figure(figsize=(8, 5))
plt.pie(frequencies_relative, labels=df['Intervalo'], autopct='%1.1f%%', startangle=140)
plt.title('Diagrama de Pastel de Frecuencia Relativa')
plt.tight_layout()
plt.show()
```

```python
# 5. Ojiva de Frecuencia Acumulada
plt.figure(figsize=(8, 5))
plt.plot(df['Intervalo'], df['Frecuencia Acumulada'], marker='o', linestyle='-', color='green')
plt.title('Ojiva de Frecuencia Acumulada')
plt.xlabel('Intervalo')
plt.ylabel('Frecuencia Acumulada')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```
